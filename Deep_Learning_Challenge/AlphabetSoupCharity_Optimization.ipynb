## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df_opti = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df_opti.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
# I would like to drop EIN, NAME, STATUS, SPECIAL_CONSIDERATIONS
app_clean_opti = application_df_opti.drop(['EIN', 'NAME', 'STATUS', 'SPECIAL_CONSIDERATIONS'], axis=1)
```

```python
# Look at APPLICATION_TYPE value counts for binning
unique_value_counts_opti = app_clean_opti .nunique()
print(unique_value_counts_opti)
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`

application_types_to_replace_opti = list(unique_value_counts_opti[unique_value_counts_opti < 500].index)

# Replace in dataframe
for app in application_types_to_replace_opti:
   application_df_opti['APPLICATION_TYPE'] = application_df_opti['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df_opti['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
CLASSIFICATION_count_opti = application_df_opti['CLASSIFICATION'].value_counts()
CLASSIFICATION_count_opti
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
CLASSIFICATION_count_opti[CLASSIFICATION_count_opti>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace_opti = list(CLASSIFICATION_count_opti[CLASSIFICATION_count_opti < 1000].index)

# Replace in dataframe
for cls in classifications_to_replace_opti:
   application_df_opti['CLASSIFICATION'] = application_df_opti['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df_opti['CLASSIFICATION'].value_counts()
```

```python
cat_vars = list(app_clean_opti.dtypes[app_clean_opti.dtypes == 'object'].index)
cat_vars
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_clean_opti = pd.get_dummies(app_clean_opti)
app_clean_opti.head()
```

```python
# Split our preprocessed data into our features and target arrays
x_opti = app_clean_opti.drop(columns=['IS_SUCCESSFUL']).values
y_opti = app_clean_opti['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset
x_train, x_test, y_train, y_test = train_test_split(x_opti, y_opti)
```

```python
# Create a StandardScaler instances
scaler_opti = StandardScaler()

# Fit the StandardScaler
X_scaler_opti = scaler_opti.fit(x_train)

# Scale the data
X_train_scaled = X_scaler_opti.transform(x_train)
X_test_scaled = X_scaler_opti.transform(x_test)
```

```python
X_test_scaled
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_features = len(x_train[0])
layer_1 = 300
layer_2 = 200
layer_3 = 100
layer_4 = 50
layer_5 = 10

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_1,
                          input_dim=number_features,
                          activation='relu')
)

# Second hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_2,
                          input_dim=layer_1,
                          activation='sigmoid')
)

# Third hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_3,
                          input_dim=layer_2,
                          activation='sigmoid')
)

# Fourth hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_4,
                          input_dim=layer_3,
                          activation='sigmoid')
)

# Fifth hidden layer
nn.add(
    tf.keras.layers.Dense(units=layer_5,
                          input_dim=layer_4,
                          activation='sigmoid')
)

# Output layer
nn.add(
    tf.keras.layers.Dense(units=1,
                          input_dim=layer_5,
                          activation='sigmoid')
)

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
